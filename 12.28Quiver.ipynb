```python
# 1简单的均匀向量场，其中所有向量指向同一方向，使用 quiver 函数绘制箭头以表示向量。这有助于理解矢量图的基本用法
import numpy as np
import matplotlib.pyplot as plt
X,Y=np.meshgrid(np.arange(-2,2,0.5),np.arange(-2,2,0.5))
U=np.ones_like(X)
V=np.zeros_like(Y)
fig, ax=plt.subplots()
ax.quiver(X,Y,U,V,scale=5,color='b')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Basic Uniform Vector Field with Arrows')
plt.show()
```

```python
# 2模拟一个漩涡场（旋转向量场），使用箭头表示向量的方向和幅度。这展示了如何通过数学函数定义复杂的向量场。
X,Y=np.meshgrid(np.arange(-3,3,0.5),np.arange(-3,3,0.5))
U=-Y
V=X
fig, ax=plt.subplots()
ax.quiver(X,Y,U,V,angles='xy',scale_units='xy',scale=1,color='r',pivot='mid')      # angles='xy'基于图中的坐标系
ax.set_xlim(-3,3)
ax.set_ylim(-3,3)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Vortex Vector Field with Arrows')
plt.show()
```

```python
# 3此示例展示了一个鞍点向量场，使用箭头表示发散和收敛的向量。这有助于学习如何可视化不稳定平衡点。
X,Y=np.meshgrid(np.arange(-2,2,0.25),np.arange(-2,2,0.25))
U=X
V=-Y
fig, ax=plt.subplots()
speed=np.sqrt(U**2+V**2)
q=ax.quiver(X,Y,U,V,speed,pivot='mid',cmap='viridis')
fig.colorbar(q,ax=ax,label='Vector Magnitude')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Saddle Point Vector Field with Arrows')
plt.show()
```

```python
# 结合正弦函数创建一个波浪状向量场，并调整箭头宽度以反映向量强度。
X,Y=np.meshgrid(np.arange(0,10,0.5),np.arange(0,10,0.5))
U=np.sin(X)
V=np.cos(Y)
fig, ax=plt.subplots(figsize=(8,8))
speed=np.sqrt(U**2+V**2)
q=ax.quiver(X,Y,U,V,speed,cmap='ocean',headwidth=4,pivot='mid',scale=15)
plt.colorbar(q,ax=ax,label='Wave Intensity')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Wave-like Vector Field with Variable Arrow Width')
plt.show()
```

```python
# 3D 螺旋场，模拟一个向上盘旋的“气旋”流场
X,Y,Z=np.meshgrid(np.linspace(-2,2,10),np.linspace(-2,2,10),np.linspace(0,5,10))
U=-Y
V=X
W=0.5*Z
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111,projection='3d')
q=ax.quiver(X,Y,Z,U,V,W,length=0.2,cmap='jet',alpha=0.8)
ax.set_xlabel('X (m)')
ax.set_ylabel('Y (m)')
ax.set_zlabel('Height (m)')
ax.set_title('3D Atmospheric Vortex Simulation')
plt.show()
```
