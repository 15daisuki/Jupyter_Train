```python
# Finite Volume Method-page 39
# 

import numpy as np
import matplotlib.pyplot as plt

# -----------------数值解  numerical solution
Phi_A = 1
Phi_B = 0

A = np.array([[1.55, -0.45, 0, 0, 0],
              [-0.55, 1, -0.45, 0, 0],
              [0, -0.55, 1, -0.45, 0],
              [0, 0, -0.55, 1, -0.45],
              [0, 0, 0, -0.55, 1.45]])

b = np.array([1.1 * Phi_A, 0, 0, 0, 0.95 * Phi_B])

Phi = np.linalg.solve(A, b)

for i, t in enumerate(Phi, 1):
    print(f"Phi{i} = {t:.3f}")

x = np.linspace(0, 1, 7) # 加上边界2点

phi_full = np.concatenate(([Phi_A], Phi, [Phi_B]))
plt.figure(figsize=(8, 6))
plt.plot(x, phi_full)
plt.show()


# ------------------解析解  analytical solution





```
