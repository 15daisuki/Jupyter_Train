```python
import numpy as np
import matplotlib.pyplot as plt
# --------------------- 示例1：基本等值线图（仅线条） ---------------------
x=np.linspace(-3,3,200)        # np.linspace(start, stop, num)
y=np.linspace(-3,3,200)        # 点
X,Y=np.meshgrid(x,y)           # 从点到面的演变 拉伸为矩阵
Z=np.sin(X*Y)*np.exp(-(X**2+Y**2)/2)
plt.figure(figsize=(8,6))      # 宽和高
contour1=plt.contour(X,Y,Z)
plt.clabel(contour1,inline=True,fontsize=10)  # plt.clabel() 给等高线标注数值,inline=True 标签嵌入到等高线中,fontsize=10 字体大小
plt.title('contour-1')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.colorbar(contour1)
plt.show()
```

```python
# --------------------- 示例2：指定等值线水平数 ---------------------
plt.figure(figsize=(8,6))
contour2=plt.contour(X,Y,Z,levels=10,colors='black') # 指定10条等值线 黑色
plt.clabel(contour2,inline=True,fontsize=9,fmt='%1.1f')
plt.title('contour-2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
# --------------------- 示例3：指定具体等值线水平值 ---------------------
levels_specific=[-0.1,-0.05,0,0.05,0.1]
plt.figure(figsize=(8,6))
contour3=plt.contour(X,Y,Z,levels=levels_specific,linewidths=2)
plt.clabel(contour3,inline=True,fontsize=10,colors='red')
plt.title('contour-3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.colorbar(contour3)
plt.show()
```

```python
# --------------------- 示例4：填充等值线图（contourf） ---------------------
plt.figure(figsize=(8,6))
contour4=plt.contourf(X,Y,Z,levels=20,cmap='viridis') # 颜色映射（colormap）
plt.title('contour-4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar(contour4,label='Z-axis') # 颜色条名字
plt.show()
```

```python
# --------------------- 示例5：线条 + 填充 + 标签（常用组合） ---------------------
plt.figure(figsize=(8,6))
contourf5=plt.contourf(X,Y,Z,levels=15,cmap='viridis',alpha=0.8)         # 填充等高线（底图）
contour5=plt.contour(X,Y,Z,levels=15,colors='white',linewidths=0.5)       # 等高线（线条层）
plt.clabel(contour5,inline=True,fontsize=8,colors='white',fmt='%1.2f')     # 等高线数值标签
plt.title('contour-5')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar(contourf5,label='Z-axis') # 颜色条名字
plt.show()
```

```python
# --------------------- 示例6：负值使用虚线 ---------------------
plt.figure(figsize=(8,6))
contour6=plt.contour(X,Y,Z,levels=10,colors='black',
                     linestyles=['dashed' if lvl<0 else 'solid' for lvl in plt.contour(X,Y,Z,levels=15).levels])
plt.clabel(contour6,inline=True,fontsize=5)            # plt.contour() 等值线标注 inline=True 嵌入
plt.title('contour-6')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
# --------------------- 示例7：不同颜色映射（cmap）对比 ---------------------
cmaps=['viridis','plasma','inferno','cividis','terrain']               # 定义方案 case1~case5
fig,axs=plt.subplots(1,len(cmaps),figsize=(20,4))                      # 1行5列 总大小20-4  len(cmaps) is column number
for i,cmap in enumerate(cmaps):                                        # 在第 i 个子图上绘制填充等高线图
    axs[i].contourf(X,Y,Z,levels=12,cmap=cmap)
    axs[i].set_title(f'color_type={cmap}')
    axs[i].set_xlabel('X')
    axs[i].set_ylabel('Y')
    plt.colorbar(axs[i].collections[0],ax=axs[i])                      # ax=axs[i] 定位colorbar
plt.suptitle('contour-7')
plt.tight_layout()                                                     # 自动几何调整
plt.show()
```
