```python
# 1基本均匀向量场流线图
import numpy as np
import matplotlib.pyplot as plt
X,Y=np.meshgrid(np.arange(-2,2,0.25),np.arange(-2,2,0.25)) # 2D网格
U=np.ones_like(X)                                # 与X形状相同的全1数组
V=np.zeros_like(Y)                               # 与Y形状相同的全0数组
fig, ax=plt.subplots()                           # 创建一个包含画布（Figure）和子图（Axes）的对象
ax.streamplot(X,Y,U,V,linewidth=1,density=1)     # ax.streamplot(...) 绘制二维流线的核心函数。
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Basic Uniform Vector Field')
plt.show()
```

```python
# 2模拟一个漩涡场（旋转向量场），类似于流体力学中的涡流
X,Y=np.meshgrid(np.arange(-3,3,0.2),np.arange(-3,3,0.2))
U=-Y
V=X
fig, ax=plt.subplots()
ax.streamplot(X,Y,U,V,color='b',arrowsize=1,density=2)    # arrowsize箭头
ax.set_xlim(-3,3)                                         # X轴范围
ax.set_ylim(-3,3)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Vortex Vector Field')
plt.show()
```

```python
# 3鞍点（saddle point）向量场，常用于描述不稳定平衡点,表示发散和收敛的流线。
X,Y=np.meshgrid(np.arange(-2,2,0.1),np.arange(-2,2,0.1))
U=X
V=-Y
fig, ax=plt.subplots()
speed=np.sqrt(U**2+V**2)                                # 计算每个网格点上的速度矢量模长（Speed）
ax.streamplot(X,Y,U,V,linewidth=0.5,cmap='viridis',density=1.5,color=speed)
plt.colorbar(ax.collections[0],ax=ax,label='Speed')    # ax.collections[0]流线的颜色映射信息
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Saddle Point Vector Field')
plt.show()
```

```python
# 4结合正弦函数创建一个波浪状向量场，并调整箭头和线宽以增强可视化
X,Y=np.meshgrid(np.arange(0,10,0.5),np.arange(0,10,0.5))
U=np.sin(X)                                                  # 水平速度随X坐标呈正弦波动。
V=np.cos(Y)
fig, ax=plt.subplots()
speed=np.sqrt(U**2+V**2)
lw=2*speed+0.5                                               # 动态线-宽线宽随速度成正比变化 最小为0.5
ax.streamplot(X,Y,U,V,broken_streamlines=False,arrowstyle='->',linewidth=lw,density=1)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Wave-like Vector Field with Variable Linewidth')
plt.show()
```

```python
# 5在多个子图中比较不同密度参数的影响，便于理解密度对流线数量的控制
X,Y=np.meshgrid(np.arange(-1,1,0.1),np.arange(-1,1,0.1))
U=Y
V=-X
fig, axs=plt.subplots(1,2,figsize=(10,5))                           # 创建一行两列的子图布局。
axs[0].streamplot(X,Y,U,V,density=0.5)                              # [0],[1]访问子图对象列表。
axs[0].set_title('Low Density(0.5)')
axs[1].streamplot(X,Y,U,V,density=3)
axs[1].set_title('High Density(3)')
for ax in axs:                                                      # 给每个子层XY轴取名-循环遍历子图列表
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
plt.tight_layout()                                                  # 自动调整子图参数
plt.show()
```
