```python
import numpy as np

# page 19
# 1D steady heat

Ta = 100
Tb = 500
A = np.array([[300, -100, 0, 0, 0],
              [-100, 200, -100, 0, 0],
              [0, -100, 200, -100, 0],
              [0, 0, -100, 200, -100],
              [0, 0, 0, -100, 300]], dtype=float) # dtype=float保证精度

b = np.array([200*Ta, 0, 0, 0, 200*Tb])


T = np.linalg.solve(A, b)  # 直接求解 Ax = b

print("各节点温度")
for i, t in enumerate(T, 1):      # enumerate(可迭代对象, start=0) 同时返回符号和元素
    print(f"T{i} = {t:.2f} °C")   # {t:.2f} 保留小数点后两位


# page 30
# 2D steady heat

A = np.array([[20, -10, 0, 0, -10, 0, 0, 0, 0, 0, 0, 0],
      [-10, 30, -10, 0, 0, -10, 0, 0, 0, 0, 0, 0],
      [0, -10, 30, -10, 0, 0, -10, 0, 0, 0, 0, 0],
      [0, 0, -10, 40, 0, 0, 0, -10, 0, 0, 0, 0],
      [-10, 0, 0, 0, 30, -10, 0, 0, -10, 0, 0, 0],
      [0, -10, 0, 0, -10, 40, -10, 0, 0, -10, 0, 0],
      [0, 0, -10, 0, 0, -10, 40, -10, 0, 0, -10, 0],
      [0, 0, 0, -10, 0, 0, -10, 50, 0, 0, 0, -10],
      [0, 0, 0, 0, -10, 0, 0, 0, 20, -10, 0, 0],
      [0, 0, 0, 0, 0, -10, 0, 0, -10, 30, -10, 0],
      [0, 0, 0, 0, 0, 0, -10, 0, 0, -10, 30, -10],
      [0, 0, 0, 0, 0, 0, 0, -10, 0, 0, -10, 40]], dtype=float)

b = np.array([500, 500, 500, 2500, 0, 0, 0, 2000, 0, 0, 0, 2000])

T = np.linalg.solve(A,b)

print("1~12节点温度")

for i, t in enumerate(T, 1):
    print(f"{i} = {t:.1f} °C")
```
