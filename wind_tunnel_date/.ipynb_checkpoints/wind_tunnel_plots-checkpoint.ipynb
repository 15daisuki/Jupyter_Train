# 風洞実験データ可視化
## Plane average 40mm N (λp=0.40)

- **18 cases**: F0, F7, F15, F30, F2H, P15, P30, SA15, SA30, SD15, SD30（左の棒グラフは0°のもの）
- **右の折れ線グラフ**: 各角度（0°, 15°, 30°, 45°, 60°, 75°, 90°）での平均風速比
- **データ**: 各CSVのB3:Q3（16点平均）→ `Vave / Vave_F0(0.25_0°)`

CSVファイルは **このノートブックと同じフォルダ** に置いてください。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import os
import warnings
warnings.filterwarnings('ignore')

# ── 設定 ────────────────────────────────────────────────────────────
DATA_DIR = '.'          # CSVが置かれているフォルダ（同じ場所なら '.'）
ANGLES   = [0, 15, 30, 45, 60, 75, 90]

# 全18ケース（グラフの凡例順）
CASES = ['F0', 'F7', 'F15', 'F30', 'F2H', 'P15', 'P30',
         'SA15', 'SA30', 'SD15', 'SD30']

# 折れ線グラフ用の色（画像に合わせた配色）
LINE_COLORS = {
    'F0' : '#C8A000',   # gold
    'F7' : '#808080',   # gray
    'F15': '#E06010',   # orange-red
    'F30': '#1060C0',   # dark blue
    'F2H': '#40B8F0',   # sky blue
    'P15': '#203080',   # navy
    'P30': '#40A040',   # green
    'SA15':'#C03020',   # dark red
    'SA30':'#404040',   # charcoal
    'SD15':'#804010',   # brown
    'SD30':'#303030',   # near-black
}

# ── CSVを読み込んで B3:Q3（行index=2, 列B〜Q）の平均を返す ─────────
def read_vave(case, angle):
    """
    ファイル名パターン: {CASE}_N_{ANGLE}.csv
    B3:Q3 → pandas では header なし読み込みで row=2, col 1〜16 (0-indexed)
    日本語環境の Excel 保存 CSV は Shift-JIS (cp932) が多い。
    """
    filename = f"{case}_N_{angle}.csv"
    filepath = os.path.join(DATA_DIR, filename)
    
    if not os.path.exists(filepath):
        return np.nan
    
    # Shift-JIS → UTF-8 の順で試みる
    for enc in ['cp932', 'shift_jis', 'utf-8', 'utf-8-sig']:
        try:
            df = pd.read_csv(filepath, header=None, encoding=enc)
            # B3:Q3 → row index 2 (3行目), column index 1〜16 (B〜Q)
            row = df.iloc[2, 1:17].astype(float)
            return row.mean()
        except Exception:
            continue
    
    print(f"[警告] 読み込み失敗: {filename}")
    return np.nan

# ── 全ケース × 全角度のデータを収集 ─────────────────────────────────
data = {}   # data[case][angle] = Vave
for case in CASES:
    data[case] = {}
    for angle in ANGLES:
        data[case][angle] = read_vave(case, angle)

# 基準値: F0 の 0° → Vave_F0(0.25_0°)
ref = data['F0'][0]
print(f"基準値 Vave_F0(0°) = {ref:.6f}")

# 正規化
ratio = {}
for case in CASES:
    ratio[case] = {a: data[case][a] / ref for a in ANGLES}

print("\n--- 正規化済み Vave / Vave_F0(0.25_0°) ---")
df_ratio = pd.DataFrame(ratio).T
df_ratio.columns = [f"{a}°" for a in ANGLES]
print(df_ratio.round(3))
```

```python
# ── 描画 ──────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.patch.set_facecolor('white')
title_str = 'Plane average 40mm N  (λp=0.40)'

# ── 左: 棒グラフ（0° のみ）────────────────────────────────────────
ax1 = axes[0]
vals_0 = [ratio[c][0] for c in CASES]
x = np.arange(len(CASES))

bars = ax1.bar(x, vals_0, color='#4472C4', width=0.6)
ax1.set_title(f'{title_str}  0°', fontsize=11)
ax1.set_ylabel('Vave / Vave_F0(0.25_0°)', fontsize=9)
ax1.set_xticks(x)
ax1.set_xticklabels(CASES, fontsize=8)
ax1.set_ylim(0, 1.8)
ax1.yaxis.set_major_locator(ticker.MultipleLocator(0.2))
ax1.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.7)
ax1.set_facecolor('white')
for spine in ax1.spines.values():
    spine.set_linewidth(0.8)

# ── 右: 折れ線グラフ（全角度）────────────────────────────────────
ax2 = axes[1]
markers = ['o','s','D','^','v','<','>','p','*','h','H']

for i, case in enumerate(CASES):
    yvals = [ratio[case][a] for a in ANGLES]
    ax2.plot(ANGLES, yvals,
             color=LINE_COLORS[case],
             marker=markers[i % len(markers)],
             markersize=5,
             linewidth=1.5,
             label=case)

ax2.set_title(title_str, fontsize=11)
ax2.set_ylabel('Vave / Vave_F0(0.25_0°)', fontsize=9)
ax2.set_xlabel('Angle (°)', fontsize=9)
ax2.set_xlim(-2, 92)
ax2.set_xticks(ANGLES)
ax2.set_ylim(0, 2.5)
ax2.yaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax2.grid(linestyle='--', linewidth=0.5, alpha=0.7)
ax2.set_facecolor('white')
for spine in ax2.spines.values():
    spine.set_linewidth(0.8)

# 凡例を2列で右上に
ax2.legend(ncol=3, fontsize=7.5, loc='upper right',
           framealpha=0.9, edgecolor='gray')

plt.tight_layout(pad=2.0)
plt.savefig('wind_tunnel_result.png', dpi=150, bbox_inches='tight')
plt.show()
print("→ wind_tunnel_result.png に保存しました")
```
